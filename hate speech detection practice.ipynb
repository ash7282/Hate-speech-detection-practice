# Import libraries
import re
import string
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score

# Define a function to clean the text
def clean(text):
    text = str(text).lower() # Convert to lowercase
    text = re.sub('https?://\S+|www\.\S+', '', text) # Remove URLs
    text = re.sub('<.*?>+', '', text) # Remove HTML tags
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text) # Remove punctuation
    text = re.sub('\n', '', text) # Remove newlines
    text = re.sub('\w*\d\w*', '', text) # Remove words containing numbers
    text = [word for word in text.split(' ') if word not in stopwords.words('english')] # Remove stopwords
    text = " ".join(text) # Join the words
    stemmer = PorterStemmer() # Initialize a stemmer
    text = [stemmer.stem(word) for word in text.split(' ')] # Stem the words
    text = " ".join(text) # Join the words
    return text

# Load the data
data = pd.read_csv('hate_speech_data.csv') # Replace with your data file
data['tweet'] = data['tweet'].apply(clean) # Clean the tweets

# Split the data into train and test sets
train_data = data.sample(frac=0.8, random_state=42) # Use 80% of the data for training
test_data = data.drop(train_data.index) # Use the remaining 20% for testing

# Vectorize the tweets using TF-IDF
vectorizer = TfidfVectorizer(max_features=1000) # Use the top 1000 words as features
X_train = vectorizer.fit_transform(train_data['tweet']) # Fit and transform the train data
X_test = vectorizer.transform(test_data['tweet']) # Transform the test data
y_train = train_data['label'] # Get the train labels
y_test = test_data['label'] # Get the test labels

# Train a logistic regression classifier
classifier = LogisticRegression(random_state=42) # Initialize a classifier
classifier.fit(X_train, y_train) # Fit the classifier on the train data

# Make predictions on the test data
y_pred = classifier.predict(X_test) # Predict the test labels

# Evaluate the performance of the classifier
acc = accuracy_score(y_test, y_pred) # Compute the accuracy score
f1 = f1_score(y_test, y_pred) # Compute the F1 score
print(f'Accuracy: {acc:.2f}')
print(f'F1 score: {f1:.2f}')
